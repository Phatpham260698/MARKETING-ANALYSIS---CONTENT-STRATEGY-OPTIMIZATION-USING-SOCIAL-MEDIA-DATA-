5. DEVELOPING MEASUREMENT METHODS AND ACCOMPANIED METRICS
5.1. Measurement methods:  
    - This report was analyzed by Funnel Analysis with these phases: Distribution -> Attention -> Interest -> Intent -> Decision   
5.1.1. Distribution:  
    - Question 1: Are the posts uploaded on platforms?  
    - This is the stage where posts are displayed on platforms, shown through Impression (The number of times the post is displayed on the viewer's screen). Although the result of the conversion is not clearly shown at this stage, it serves as the benchmark for most of the metrics   
    - Metric: Impression - Although this metric only shows the number of posts displayed and doesn't measure effectiveness, it's the input to the entire funnel (the common denominator of most metrics).  
5.1.2. Attention: 
    - Question 2: Do Viewers actually view posts?
    - This is the stage where Viewers actually view posts beyond the 'skimming' phase and absorb the content of a post in actual. To measure accurately at this stage, it should be further defined: 'How long must viewers remain on the post to be considered as having actually viewed the content?' - something that the provided data does not reflect yet.   
    - Metrics: View_Rate   
5.1.3. Interest: 
    - Question 3: Are posts interacted and the interactions valued?  
    - At this stage, posts are evaluated based on the reactions of viewers, and all of the interactions (e.g, Like, Share, Comment,...) are equally considered. However, it cannot be considered the potential of spreading on platforms as well as the valued interactions through its volumes, so it needs a different metric, with a different proportion of interactions, which means that the size is directly proportional to the extent of spread.  
    - Metrics: Engagement_Rate, Weight_Engagement_Rate, Share_Rate  
5.1.4. Intent:  
    - Question 4: Does the content encourage action?/Does interactive content encourage action?    
    - Intent stage - It is shown that Viewers actually click on posts, especially for advertising/sales posts, demonstrating the viewer's desire to learn more about products.  
    - Metrics: CTR, Engagement_To_Click  
5.1.5. Decision: 
    - Question 5: Is the content effective overall?  
    - Actually, this stage isn't strictly part of the Behaviour Analysis, but it can be added to the analysis process for the overall effectiveness of a whole series of Viewer behaviours, while there a multiple metrics in this report. In this analysis, CPI is composed of three sub-metrics: Engagement_Rate, CTR, and View_Rate, with weights of 0.4, 0.3, and 0.3 respectively. Within this metric, Engagement_Rate will be chosen as the more important (though not significantly different) metric, focusing primarily on the level of Viewer interaction with the post.  
    - Metric: CPI  
5.1.6. Additional metrics:  
    - In addition to metrics built based on funnel analysis, secondary metrics related to posting time can also provide insight into the optimal timeframe for efficiency: Time_Slot_Efficiency.   
5.2. CONSTRUCT DETAILED METRICS:  
    - View_Rate = View/Impressions 
    - Engagement_Rate = Engagement/Impressions  
    - Weighted_Engagement_Rate = (Likes + 2 * Comments + 3 * Shares)/Impressions  
    - Share_Rate = Shares / Impressions: tiềm năng lan tỏa/tăng tương tác  
    - Click Through Rate (CTR) = Clicks / Impressions 
    - Engagement_to_Click = Clicks / Engagement   
    - CPI = 0.4 * Engagement_Rate + 0.3 * CTR + 0.3 * View_Rate   
    - Time_Slot_Efficiency = Avg(Engagement_Rate) theo Post_Hour  

```python
#Create functions 
def group_and_metrics(df, group_col):
    grouped = (
        df.groupby(['Platform',group_col],as_index=False,observed=True)
        .agg({
            "Impressions": "sum",
            "Engagement": "sum",
            "Clicks": "sum",
            "Shares": "sum",
            "Likes": "sum",
            "Comments": "sum",
            "Views": "sum"
        })
    )
    
    grouped["View_Rate"] = grouped["Views"].div(grouped["Impressions"]).fillna(0)* 100
    grouped["Engagement_Rate"] = grouped["Engagement"].div(grouped["Impressions"]).fillna(0)* 100
    grouped["Weighted_Engagement_Rate"] = (grouped["Likes"]*1 + grouped["Comments"]*2 + grouped["Shares"]*3).div(grouped["Impressions"]).fillna(0)*100
    grouped["Share_Rate"] = grouped["Shares"].div(grouped["Impressions"]).fillna(0)* 100
    grouped["CTR"] = grouped["Clicks"].div(grouped["Impressions"]).fillna(0)* 100 
    grouped["Engagement_to_Click"] = grouped["Clicks"].div(grouped["Engagement"]).fillna(0)*100
    grouped["CPI"] = grouped["Engagement_Rate"]*0.4 + grouped["CTR"]*0.3 + grouped["View_Rate"]*0.3

    a = grouped[['Platform',group_col,'View_Rate','Engagement_Rate','Weighted_Engagement_Rate','Share_Rate','CTR','Engagement_to_Click','CPI']]
    
    # Show selected columns   
    return a

def group_and_metrics2(df, group_col):
    grouped = (
        df.groupby([group_col],as_index=False,observed=True)
        .agg({
            "Impressions": "sum",
            "Engagement": "sum",
            "Clicks": "sum",
            "Shares": "sum",
            "Likes": "sum",
            "Comments": "sum",
            "Views": "sum"
        })
    )
    
    grouped["View_Rate"] = grouped["Views"].div(grouped["Impressions"]).fillna(0)* 100
    grouped["Engagement_Rate"] = grouped["Engagement"].div(grouped["Impressions"]).fillna(0)* 100
    grouped["Weighted_Engagement_Rate"] = (grouped["Likes"]*1 + grouped["Comments"]*2 + grouped["Shares"]*3).div(grouped["Impressions"]).fillna(0)*100
    grouped["Share_Rate"] = grouped["Shares"].div(grouped["Impressions"]).fillna(0)* 100
    grouped["CTR"] = grouped["Clicks"].div(grouped["Impressions"]).fillna(0)* 100 
    grouped["Engagement_to_Click"] = grouped["Clicks"].div(grouped["Engagement"]).fillna(0)*100
    grouped["CPI"] = grouped["Engagement_Rate"]*0.4 + grouped["CTR"]*0.3 + grouped["View_Rate"]*0.3
    
    a = grouped[[group_col,'View_Rate','Engagement_Rate','Weighted_Engagement_Rate','Share_Rate','CTR','Engagement_to_Click','CPI']]
    
    # Show selected columns      
    return a

def Time_Slot_Efficiency (df,df_col):
    grouped = (
        df.groupby([df_col], as_index=False,observed=True)
        .agg({
            "Impressions": "sum",
            "Engagement": "sum",
            "Clicks": "sum",
            "Shares": "sum",
            "Likes": "sum",
            "Comments": "sum",
            "Views": "sum"
        })
    )

    grouped["View_Rate"] = grouped["Views"].div(grouped["Impressions"]).fillna(0)* 100
    grouped["Engagement_Rate"] = grouped["Engagement"].div(grouped["Impressions"]).fillna(0)* 100
    grouped["Weighted_Engagement_Rate"] = (grouped["Likes"]*1 + grouped["Comments"]*2 + grouped["Shares"]*3).div(grouped["Impressions"]).fillna(0)*100
    grouped["Share_Rate"] = grouped["Shares"].div(grouped["Impressions"]).fillna(0)* 100
    grouped["CTR"] = grouped["Clicks"].div(grouped["Impressions"]).fillna(0)* 100 
    grouped["Engagement_to_Click"] = grouped["Clicks"].div(grouped["Engagement"]).fillna(0)*100
    grouped["CPI"] = grouped["Engagement_Rate"]*0.4 + grouped["CTR"]*0.3 + grouped["View_Rate"]*0.3

    b = grouped[[df_col,'View_Rate','Engagement_Rate','Weighted_Engagement_Rate','Share_Rate','CTR','Engagement_to_Click','CPI']]
    
    # Show selected columns 
    return b

def metrics_and_area(df, group_col):
    grouped = (
        df.groupby([group_col,'Content_Category'],as_index=False,observed=True)
        .agg({
            "Impressions": "sum",
            "Engagement": "sum",
            "Clicks": "sum",
            "Shares": "sum",
            "Likes": "sum",
            "Comments": "sum",
            "Views": "sum"
        })
    )
    
    grouped["View_Rate"] = grouped["Views"].div(grouped["Impressions"]).fillna(0)* 100
    grouped["Engagement_Rate"] = grouped["Engagement"].div(grouped["Impressions"]).fillna(0)* 100
    grouped["Weighted_Engagement_Rate"] = (grouped["Likes"]*1 + grouped["Comments"]*2 + grouped["Shares"]*3).div(grouped["Impressions"]).fillna(0)*100
    grouped["Share_Rate"] = grouped["Shares"].div(grouped["Impressions"]).fillna(0)* 100
    grouped["CTR"] = grouped["Clicks"].div(grouped["Impressions"]).fillna(0)* 100 
    grouped["Engagement_to_Click"] = grouped["Clicks"].div(grouped["Engagement"]).fillna(0)*100
    grouped["CPI"] = grouped["Engagement_Rate"]*0.4 + grouped["CTR"]*0.3 + grouped["View_Rate"]*0.3
    
    a = grouped[[group_col,'Content_Category','View_Rate','Engagement_Rate','Weighted_Engagement_Rate','Share_Rate','CTR','Engagement_to_Click','CPI']]
    
    # Show selected columns     
    return a
```

10. WHICH IS THE EFFECTIVE HASHTAG IN INCREASING VIEW TIMES OR CLICK?  
- The 5 hashtag '#ProductDemo', '#CustomerStory','#TrendingNow','#SuccessStory' account for 80% impressions  
- The 7 hashtag #ProductDemo,#CustomerStory,#TrendingNow,#SuccessStory,#SaaSLaunch,#DidYouKnow,#FeatureHighlight account for 80% clicks  
=> It should be focused on developing these post having contents and main hashtag among hashtags above to optimize the number of views and clicks.  


```python
a = df_dataset.groupby(['Main_Hashtag']).agg({
    'Impressions' : 'sum'
}).reset_index().sort_values(by='Impressions',ascending=False)

b = df_dataset.groupby(['Main_Hashtag']).agg({
    'Clicks':'sum'
}).reset_index().sort_values(by='Clicks',ascending=False)

#Sum Accumulate Impressions and Clicks
a['Impressions_accumulate'] = a['Impressions'].cumsum()

b['Clicks_accumulate'] = b['Clicks'].cumsum()

#Scalar Impressions and Clicks
Impressions_Scalar = int (a['Impressions'].sum())
Clicks_Scalar = int (b['Clicks'].sum())

#Pareto analysis
a['Impressions_Pareto_Analysis'] = a['Impressions_accumulate']/Impressions_Scalar
b['Clicks_Pareto_Analysis'] = b['Clicks_accumulate']/Clicks_Scalar

#Plot
fig, (ax1, ax2) = plt.subplots(2,1, figsize=(10,8))

# ======================
# IMPRESSIONS
# ======================
ax1.bar(a['Main_Hashtag'], a['Impressions'])
ax1.set_ylabel('Impressions')
ax1.set_title('Tác động của Hashtag tới Impressions')

ax1_twin = ax1.twinx()
ax1_twin.plot(a['Main_Hashtag'], a['Impressions_Pareto_Analysis'], marker='o', color='green')
ax1_twin.set_ylabel('Cumulative %')
ax1_twin.axhline(0.8, linestyle='--', color='gray')

# rotate label ax1
ax1.tick_params(axis='x', rotation=45)

# ======================
# CLICKS
# ======================
ax2.bar(b['Main_Hashtag'], b['Clicks'])
ax2.set_ylabel('Clicks')
ax2.set_title('Tác động của Hashtag tới Clicks')

ax2_twin = ax2.twinx()
ax2_twin.plot(b['Main_Hashtag'], b['Clicks_Pareto_Analysis'], marker='o', color='green')
ax2_twin.set_ylabel('Cumulative %')
ax2_twin.axhline(0.8, linestyle='--', color='gray')

# rotate label ax2
ax2.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
