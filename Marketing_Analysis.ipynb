5. XÂY DỰNG PHƯƠNG PHÁP ĐO LƯỜNG VÀ METRICS ĐI KÈM  
5.1. PHƯƠNG PHÁP ĐO LƯỜNG:  
    - Thực hiện đo lường theo Funnel Analysis như sau: Distribution -> Attention -> Interest -> Intent -> Decision   
5.1.1. Distribution:  
    - Trả lời cho câu hỏi: Bài đăng có được phân phối lên các nền tảng không?  
    - Đây là giai đoạn bài đăng được đưa lên các nền tảng/màn hình người xem, Được thể hiện thông qua Impression (Số lần bài đăng được hiển thị), tuy tại bước này chưa thể hiện rõ kết quả chuyển đổi của Viewer nhưng đây là mẫu số cho hầu hết các metric đo lường  
    - Các metric đo lường: Impression - chỉ số này tuy chỉ thể hiện số lượng bài đăng được hiển thị, không đo lường được tính hiệu quả nhưng là đầu vào của toàn bộ funnel (mẫu số chung của hầu hết các metric)  
5.1.2. Attention: 
    - Trả lời cho câu hỏi: Người dùng có có thực sự xem bài đăng?
    - Đây là giai đoạn Viewer thực sự xem bài đăng, vượt qua giai đoạn chỉ "lướt" qua và thực sự tiếp nhận nội dung của bài đăng, tại bước này để đo lường chính xác, cần định nghĩa thêm 'Viewer dừng lại bao lâu để được xem là thực sự xem nội dung' - điều này ở dữ liệu được cung cấp chưa thể hiện được  
    - Các metric đo lường: View_Rate   
5.1.3. Interest: 
    - Trả lời cho câu hỏi: Bài đăng có được tương tác và tương tác có "giá trị" không?  
    - Tại bước này, bài đăng được đánh giá tạo ra được phản ứng của Viewer hay không, thước đo được đo lường một cách tổng thể, các loại tương tác được xem xét với mức độ ngang nhau. Tuy nhiên, nếu chỉ đo lường đến mức độ như trên thì không thể đánh giá được tiềm năng lan truyền trên các nền tảng cũng như mức độ tương tác có "giá trị" hay không, vì thế cần thêm một thước đo với tỷ trọng tác động của mỗi loại tương tác là khác nhau, độ lớn tỷ lệ thuận với mức độ lan truyền  
    - Các metric đo lường: Engagement_Rate, Weight_Engagement_Rate, Share_Rate  
5.1.4. Intent:  
    - Trả lời cho câu hỏi: Nội dung có thúc đẩy hành động/Nội dung được tương tác có thúc đẩy hành động  
    - Giai đoạn hành động - giai đoạn thể hiện Viewer thực hiện click vào bài đăng, thông thường đối với các bài viết dạng quảng cáo/bán hàng, thì đây là bước quan trọng, thể hiện ý chí Viewer muốn tìm hiểu thêm thông tin về sản phẩm hàng hóa  
    - Các metric đo lường: CTR, Engagement_To_Click   
5.1.5. Decision: 
    - Trả lời cho câu hỏi: Nội dung bài đăng có hiệu quả tổng thể hay không?  
    - Thực ra bước này không hẳn nằm trong hành vi của Viewer, tuy nhiên có thể thêm vào trong quá trình phân tích để đưa ra được hiệu quả tổng thể của cả một chuỗi hành vi của Viewer (khi có nhiều metric để đo lường và cần một góc nhìn hiệu quả tổng thể), trong bài phân tích này CPI được cấu thành từ 3 chỉ số con: Engagement_Rate, CTR, View_Rate với trọng số lần lượt 0.4, 0.3, 0.3. Trong chỉ số này sẽ chọn  Engagement_Rate làm chỉ số trọng yếu hơn (tuy không quá khác biệt) với việc quan tâm hơn cả là mức độ tương tác của Viewer với bài đăng  
    - Các metric đo lường: CPI  
5.1.6. Các metric đi kèm:  
    - Ngoài các metric được xây dựng dựa trên funnel analysis, các metric phụ liên quan đến thời gian đăng bài cũng có thể đưa ra góc nhìn về khung thời gian để tối ưu hiệu quả: Time_Slot_Efficiency  
5.2. XÂY DỰNG CÁC METRIC CHI TIẾT:  
    - View_Rate = View/Impressions 
    - Engagement_Rate = Engagement/Impressions  
    - Weighted_Engagement_Rate = (Likes + 2 * Comments + 3 * Shares)/Impressions  
    - Share_Rate = Shares / Impressions: tiềm năng lan tỏa/tăng tương tác  
    - Click Through Rate (CTR) = Clicks / Impressions 
    - Engagement_to_Click = Clicks / Engagement   
    - CPI = 0.4 * Engagement_Rate + 0.3 * CTR + 0.3 * View_Rate   
    - Time_Slot_Efficiency = Avg(Engagement_Rate) theo Post_Hour  

```python
#Tạo function 
def group_and_metrics(df, group_col):
    # Group theo cột được truyền vào
    grouped = (
        df.groupby(['Platform',group_col],as_index=False,observed=True)
        .agg({
            "Impressions": "sum",
            "Engagement": "sum",
            "Clicks": "sum",
            "Shares": "sum",
            "Likes": "sum",
            "Comments": "sum",
            "Views": "sum"
        })
    )
    
    # Tính các chỉ số
    grouped["View_Rate"] = grouped["Views"].div(grouped["Impressions"]).fillna(0)* 100
    grouped["Engagement_Rate"] = grouped["Engagement"].div(grouped["Impressions"]).fillna(0)* 100
    grouped["Weighted_Engagement_Rate"] = (grouped["Likes"]*1 + grouped["Comments"]*2 + grouped["Shares"]*3).div(grouped["Impressions"]).fillna(0)*100
    grouped["Share_Rate"] = grouped["Shares"].div(grouped["Impressions"]).fillna(0)* 100
    grouped["CTR"] = grouped["Clicks"].div(grouped["Impressions"]).fillna(0)* 100 
    grouped["Engagement_to_Click"] = grouped["Clicks"].div(grouped["Engagement"]).fillna(0)*100
    grouped["CPI"] = grouped["Engagement_Rate"]*0.4 + grouped["CTR"]*0.3 + grouped["View_Rate"]*0.3

    a = grouped[['Platform',group_col,'View_Rate','Engagement_Rate','Weighted_Engagement_Rate','Share_Rate','CTR','Engagement_to_Click','CPI']]
    
    # Hiển thị các chỉ số cần xem    
    return a

def group_and_metrics2(df, group_col):
    # Group theo cột được truyền vào
    grouped = (
        df.groupby([group_col],as_index=False,observed=True)
        .agg({
            "Impressions": "sum",
            "Engagement": "sum",
            "Clicks": "sum",
            "Shares": "sum",
            "Likes": "sum",
            "Comments": "sum",
            "Views": "sum"
        })
    )
    
    # Tính các chỉ số
    grouped["View_Rate"] = grouped["Views"].div(grouped["Impressions"]).fillna(0)* 100
    grouped["Engagement_Rate"] = grouped["Engagement"].div(grouped["Impressions"]).fillna(0)* 100
    grouped["Weighted_Engagement_Rate"] = (grouped["Likes"]*1 + grouped["Comments"]*2 + grouped["Shares"]*3).div(grouped["Impressions"]).fillna(0)*100
    grouped["Share_Rate"] = grouped["Shares"].div(grouped["Impressions"]).fillna(0)* 100
    grouped["CTR"] = grouped["Clicks"].div(grouped["Impressions"]).fillna(0)* 100 
    grouped["Engagement_to_Click"] = grouped["Clicks"].div(grouped["Engagement"]).fillna(0)*100
    grouped["CPI"] = grouped["Engagement_Rate"]*0.4 + grouped["CTR"]*0.3 + grouped["View_Rate"]*0.3
    
    a = grouped[[group_col,'View_Rate','Engagement_Rate','Weighted_Engagement_Rate','Share_Rate','CTR','Engagement_to_Click','CPI']]
    
    # Hiển thị các chỉ số cần xem    
    return a

def Time_Slot_Efficiency (df,df_col):
    grouped = (
        df.groupby([df_col], as_index=False,observed=True)
        .agg({
            "Impressions": "sum",
            "Engagement": "sum",
            "Clicks": "sum",
            "Shares": "sum",
            "Likes": "sum",
            "Comments": "sum",
            "Views": "sum"
        })
    )

    grouped["View_Rate"] = grouped["Views"].div(grouped["Impressions"]).fillna(0)* 100
    grouped["Engagement_Rate"] = grouped["Engagement"].div(grouped["Impressions"]).fillna(0)* 100
    grouped["Weighted_Engagement_Rate"] = (grouped["Likes"]*1 + grouped["Comments"]*2 + grouped["Shares"]*3).div(grouped["Impressions"]).fillna(0)*100
    grouped["Share_Rate"] = grouped["Shares"].div(grouped["Impressions"]).fillna(0)* 100
    grouped["CTR"] = grouped["Clicks"].div(grouped["Impressions"]).fillna(0)* 100 
    grouped["Engagement_to_Click"] = grouped["Clicks"].div(grouped["Engagement"]).fillna(0)*100
    grouped["CPI"] = grouped["Engagement_Rate"]*0.4 + grouped["CTR"]*0.3 + grouped["View_Rate"]*0.3

    b = grouped[[df_col,'View_Rate','Engagement_Rate','Weighted_Engagement_Rate','Share_Rate','CTR','Engagement_to_Click','CPI']]
    
    # Hiển thị các chỉ số cần xem
    return b

def metrics_and_area(df, group_col):
    # Group theo cột được truyền vào
    grouped = (
        df.groupby([group_col,'Content_Category'],as_index=False,observed=True)
        .agg({
            "Impressions": "sum",
            "Engagement": "sum",
            "Clicks": "sum",
            "Shares": "sum",
            "Likes": "sum",
            "Comments": "sum",
            "Views": "sum"
        })
    )
    
    # Tính các chỉ số
    grouped["View_Rate"] = grouped["Views"].div(grouped["Impressions"]).fillna(0)* 100
    grouped["Engagement_Rate"] = grouped["Engagement"].div(grouped["Impressions"]).fillna(0)* 100
    grouped["Weighted_Engagement_Rate"] = (grouped["Likes"]*1 + grouped["Comments"]*2 + grouped["Shares"]*3).div(grouped["Impressions"]).fillna(0)*100
    grouped["Share_Rate"] = grouped["Shares"].div(grouped["Impressions"]).fillna(0)* 100
    grouped["CTR"] = grouped["Clicks"].div(grouped["Impressions"]).fillna(0)* 100 
    grouped["Engagement_to_Click"] = grouped["Clicks"].div(grouped["Engagement"]).fillna(0)*100
    grouped["CPI"] = grouped["Engagement_Rate"]*0.4 + grouped["CTR"]*0.3 + grouped["View_Rate"]*0.3
    
    a = grouped[[group_col,'Content_Category','View_Rate','Engagement_Rate','Weighted_Engagement_Rate','Share_Rate','CTR','Engagement_to_Click','CPI']]
    
    # Hiển thị các chỉ số cần xem    
    return a
```
